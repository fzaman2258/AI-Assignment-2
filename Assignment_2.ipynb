# I wanted to explore how efficient certain algorithms are for finding integrals of functions

## Make sure you import the following libraries:
## - Numpy
## - Matplotlib
## - Math

## You can change the function f(x) to any of your choosing, but for this example I choose cos(x), and I want to find the integral from 0 to 2 pi of cos(x)

### Here I am defining my libraries, function, subinterval, and the array of values for x and y and using the trapezoidal rule 

```python
import numpy as np
from numpy import linspace
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import math

def f(x):
    return np.cos(x)
a, b = 0,2*np.pi
N = 10 #number of subintervals
#Trapezoidal graph
x = np.linspace(a,b,N+1)
y = f(x)
#y=f(x)
X = np.linspace(a,b,100)
Y = f(X)
fig = plt.figure(1,figsize=(10,10))
plt.title('Trapezoidal Rule y = cos(x) N={}'.format(N))
plt.plot(X,Y,alpha=1.0, color ='green')
for i in range (N):
    left_height = [x[i],x[i],x[i+1],x[i+1]]
    right_height = [0,f(x[i]),f(x[i+1]),0]
    plt.fill(left_height,right_height,color='red',edgecolor='blue')

plt.show()
fig.savefig('trap.png',dpi=500,bbox_inches='tight')
```

### I am using the Midpoint rule here, which differs from the trapezoidal rule in that we choose f((x1+x2)/2) as our value for the height. I create the list of vectors and plot it

```python
x = linspace(a,b,N+1)
y = f(x)

X = np.linspace(a,b,10*N+1)
Y = f(X)
fig = plt.figure(2,figsize=(10,10))
plt.plot(X,Y,color='blue')
x_mid = (x[:-1] + x[1:])/2
y_mid = f(x_mid)
plt.plot(x_mid,y_mid,'r.',markersize=10)
plt.bar(x_mid,y_mid,width=(b-a)/N,edgecolor='black')
plt.title('Midpoint Rule for y = cos(x) N={}'.format(N))
plt.show()

fig.savefig('midpoint.png',dpi=500,bbox_inches='tight')
```

### I am using the lefthand rule here, which differs from the midpoint rule in that we choose f(x1) as our value for the height. I create the list of vectors and plot it

```python
fig = plt.figure(3,figsize=(10,10))
plt.plot(X,Y,color='blue')
x_left = x[:-1]
y_left = y[:-1]
plt.plot(x_left,y_left,'r.',markersize=10)
plt.bar(x_left,y_left,width=(b-a)/N,align='edge',edgecolor='black')
plt.title('Lefthand Rule for y = cos(x) N={}'.format(N))
plt.show()

fig.savefig('lefthand.png',dpi=500,bbox_inches='tight')
```

### I am using the righthand rule here, which differs from the lefthand rule in that we choose f(x2) as our value for the height. I create the list of vectors and plot it

```python
fig = plt.figure(4,figsize=(10,10))
plt.plot(X,Y,color='blue')
x_right = x[1:]
y_right = y[1:]
plt.plot(x_right,y_right,'r.',markersize=10)
plt.bar(x_right,y_right,width=-(b-a)/N,align='edge',edgecolor='black')
plt.title('Righthand Rule for y = cos(x) N={}'.format(N))
plt.show()

fig.savefig('righthand.png',dpi=500,bbox_inches='tight')
```
